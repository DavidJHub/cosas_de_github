```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy import stats
import time
import progressbar
from matplotlib import cm
```

```python
def caminata(N,pos):
    steps=[]
    road=[]
    
    for i in range(N):
        r=np.random.rand()
        if r<0.5:
            steps.append(-1)
            pos=pos-1
            road.append(pos)
            #print(pos)
        else:
            steps.append(1)
            pos=pos+1
            road.append(pos)
            #print(pos)

    return np.asarray(steps),np.asarray(road)
```

```python
save=caminata(100,0)
print(save)
plt.plot(save[1])
```

```python
def calcdirec(angle):
    x=np.sin(angle)
    y=np.cos(angle)
    return x,y
```

```python
def caminata2D(N):
    
    steps=[]
    counter=[0]
    #road=[]
    xactual=0
    yactual=0
    
    X=[xactual]
    Y=[yactual]
    angles=np.linspace(0,2*np.pi,100)
    for i in range(N):
        r=random.choice(angles)
        newx=calcdirec(r)[0]
        newy=calcdirec(r)[1]
        xactual=xactual+newx
        yactual=yactual+newy
        X.append(xactual)
        Y.append(yactual)
        counter.append(i)
        
        
        
    return np.asarray(X),np.asarray(Y),np.asarray(counter)
```
